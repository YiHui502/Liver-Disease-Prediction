```python
# creating print_metrics function
def print_metrics(labels, preds):
    print("Precision Score: {}".format(precision_score(labels, preds)))
    print("Recall Score: {}".format(recall_score(labels, preds)))
    print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
    print("F1 Score: {}".format(f1_score(labels, preds)))
```

### 1. SUPPORT VECTOR MACHINES (SVM) ALGORITHM



```python
# # making copy
# liver_final = df1.copy()

# # creating train_test_splot
# X = liver_final.drop(['Liver_Disease'], axis=1)
# y = liver_final['Liver_Disease']

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# instantiating, fitting, and predicting using SVM algorithm model
svm_clf = svm.SVC(random_state=42)

svm_clf.fit(X_train, y_train)

svm_y_trials = svm_clf.predict(X_train)

svm_y_pred = svm_clf.predict(X_test)

# showing metrics for testing set
print('SVM Test Metrics:')
print('\n')
print_metrics(y_test, svm_y_pred)
print('\n')

# showing confusion matrix of SVM algorithm
print('Confusion Matrix: \n', confusion_matrix(y_test,svm_y_pred))
sns.heatmap(confusion_matrix(y_test,svm_y_pred),annot=True,fmt="d")

print()
# showing auc of SVM algorithm
svm_false_positive_rate, svm_true_positive_rate, svm_thresholds = roc_curve(y_test, svm_y_pred)
svm_auc = auc(svm_false_positive_rate, svm_true_positive_rate)
print('SVM AUC:', svm_auc)
print()

svm_acc=accuracy_score(y_test,svm_y_pred)
svm_prec=precision_score(y_test,svm_y_pred)
svm_rec=recall_score(y_test,svm_y_pred)
svm_f1=f1_score(y_test,svm_y_pred,average='macro')
svm_mas=mean_absolute_error(y_test,svm_y_pred)
svm_rms=np.sqrt(mean_squared_error(y_test,svm_y_pred))
print(classification_report(y_test,svm_y_pred ))

svm_score = round(svm_clf.score(X_train, y_train)*100,2)
svm_score_test = round(svm_clf.score(X_test, y_test)*100,2)
print('SVM Training Score: \n', svm_score)
print('SVM Test Score: \n', svm_score_test)
```
